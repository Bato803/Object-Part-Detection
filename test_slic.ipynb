```python
import skimage.segmentation as ss
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
img1 = plt.imread("./test_data/cat.jpg").reshape((1, 375, 500, 3))
img2 = plt.imread("./test_data/horse.jpg").reshape((1, 375, 500, 3))
fig = plt.figure()
a = fig.add_subplot(1,2,1)
plt.imshow(img1[0,:])
plt.axis("off")
a.set_title('cat')
a = fig.add_subplot(1,2,2)
plt.imshow(img2[0,:])
plt.axis("off")
a.set_title("horse")
```

```python
fig = plt.figure()
fig.set_figwidth(18)
fig.set_figheight(8)
fig.add_subplot(4,6,1)
plt.imshow(img1[0,:])
plt.axis("off")
fig.add_subplot(4,6,13)
plt.imshow(img2[0,:])
plt.axis("off")

for i, c in enumerate([0.01, 0.1, 1, 10, 100]):
    seg1 = ss.slic(img1[0,:], n_segments=1000, compactness=c)
    border_image1 = ss.mark_boundaries(img1[0,:], seg1, (0, 0, 0))
    fig.add_subplot(4,6,i+2,title=("compact:",c))
    plt.imshow(seg1)
    plt.axis("off")
    fig.add_subplot(4,6,i+8)
    plt.imshow(border_image1)
    plt.axis("off")
    
    seg2 = ss.slic(img2[0,:], n_segments=1000, compactness=c)
    border_image2 = ss.mark_boundaries(img2[0,:], seg2, (0, 0, 0))
    fig.add_subplot(4,6,i+14,title=("compact:",c))
    plt.imshow(seg2)
    plt.axis("off")
    fig.add_subplot(4,6,i+20)
    plt.imshow(border_image2)
    plt.axis("off")
```

```python
seg_large1 = ss.slic(img1[0,:], n_segments=100, compactness=10)
large_border1 = ss.mark_boundaries(img1[0,:], seg_large1, (0, 0, 0))

seg_large2 = ss.slic(img2[0,:], n_segments=100, compactness=10)
large_border2 = ss.mark_boundaries(img2[0,:], seg_large2, (0, 0, 0))

b = plt.figure(figsize=(18,8))
b.add_subplot(121)
plt.imshow(large_border1)
plt.axis("off")
b.add_subplot(122)
plt.imshow(large_border2)
plt.axis("off")
```
